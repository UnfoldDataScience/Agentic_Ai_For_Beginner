```python
import boto3
```

```python
import json
import uuid
session_id = str(uuid.uuid4())
```

```python
ROLE_ARN = "arn:aws:iam::442426865700:role/service-role/AmazonBedrockExecutionRoleForAgents_H8OC1GUC5CO"
REGION = 'us-east-1'
```

```python
sts_client = boto3.client('sts')
```

```python
assumed_role_object = sts_client.assume_role(
    RoleArn=ROLE_ARN, RoleSessionName='MySession'
)
credentials = assumed_role_object['Credentials']
```

```python
# Create a new client using the assumed role credentials
bedrock_agent_runtime = boto3.client(
    'bedrock-agent-runtime',  # Correct: Use 'bedrock-agent-runtime'
    aws_access_key_id=credentials['AccessKeyId'],
    aws_secret_access_key=credentials['SecretAccessKey'],
    aws_session_token=credentials['SessionToken'],
    region_name=REGION
)
```

```python
# Input for the agent
input_text = """Hi I need help with choosing a course, can you help"""

body = {
    'inputText': input_text,
}
```

```python
try:
    invoke_agent_response = bedrock_agent_runtime.invoke_agent(
        agentId='M1WVZ52PBV', agentAliasId='USMBHYCFIN', inputText=input_text, sessionId=session_id
    )
    event_stream = invoke_agent_response["completion"]
    
    for event in event_stream:
        print(event)

except Exception as e:
    print(f"Error: {e}")
```
